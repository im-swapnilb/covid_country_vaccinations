```python
sns.boxplot(cov_data['people_vaccinated'][50:100])
```

```python
sns.barplot(cov_data['country'][100:300], cov_data['total_vaccinations'])
```

```python
countries = cov_data['country'].values[20:500]
vaccines = cov_data['vaccines'].values[20:500]
plt.bar(countries,vaccines)
plt.xlabel('country')
plt.ylabel('vaccines')
plt.title('All Countries data')
plt.xticks(rotation = 100)
plt.show()
```

```python
#Number of countries using different vaccine types/combinations
cov_data.groupby(['country'],sort=False, as_index=False)['vaccines'].first()
plt.figure(figsize=(15,5))
plt.xticks(rotation=90)
sns.set(style = 'whitegrid')
sns.countplot(x="vaccines", data = cov_data.groupby(['country'],sort=False, as_index=False)['vaccines'].first())
plt.title("Countries using Different Vaccines Types\n",size =20)
plt.ylabel("No of Countries",size = 15)
plt.xlabel("\nVaccine Types\n\n",size = 15)
```
